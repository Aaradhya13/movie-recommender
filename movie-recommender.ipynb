```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    if isinstance(text, str):  # Check if the input is a string
        try:
            text = ast.literal_eval(text)  # Convert string to a Python object
        except ValueError:
            return []  # Handle malformed strings gracefully
    if isinstance(text, list):  # Ensure it's a list
        for i in text:
            if isinstance(i, dict) and 'name' in i:  # Check structure
                L.append(i['name'])
            elif isinstance(i, str):  # If it's just a string, add directly
                L.append(i)
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast

def convert_combined(text, limit=3):
    L = []
    if isinstance(text, str):  # Check if the input is a string
        try:
            text = ast.literal_eval(text)  # Convert string to a Python object
        except ValueError:
            return []  # Handle malformed strings gracefully
    if isinstance(text, list):  # Ensure it's a list
        counter = 0
        for i in text:
            if counter < limit:  # Limit the number of elements
                if isinstance(i, dict) and 'name' in i:  # Check structure
                    L.append(i['name'])
                elif isinstance(i, str):  # If it's just a string, add directly
                    L.append(i)
                counter += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_combined)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    if isinstance(text, str):  # Check if the input is a string
        try:
            text = ast.literal_eval(text)  # Convert string to a Python object
        except (ValueError, SyntaxError):  # Handle malformed strings
            return []
    if isinstance(text, list):  # Ensure the input is a list
        for i in text:
            if isinstance(i, dict) and i.get('job') == 'Director':  # Check structure
                L.append(i.get('name'))
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies['keywords'] = movies['keywords'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies['cast'] = movies['cast'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies['crew'] = movies['crew'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  # Return the joined string with spaces

# Apply the stem function to the 'tags' column
new['tags'] = new['tags'].apply(stem)
```

```python
print(new)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
ps.stem('loved')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Gandhi')
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
